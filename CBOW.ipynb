```python
class cbow(nn.Module):
    def __init__(self,vocab_size,embedding_dim,num_classes):
        super(cbow,self).__init__()
        self.embedding= nn.EmbeddingBag(vocab_size,embedding_dim,sparse=False)
        self.linear1 = nn.EmbeddingBag(embedding_dim,embedding_dim//2)
        self.fc = nn.Linear(embedding_dim//2,vocab_size)
        
    def init_weights(self):
        range_= 0.5
        self.embedding.wieght.data.uniform_(-range_,range_)
        self.fc.weight.data.uniform_(-range_,range_)
        self.fc.bais.data.zero_()
        
    def forward(self,text,offsets):
        out= self.embedding(text,offsets)
        out= torch.relu(self.linear1(out))
        return self.fc(out) 
```
